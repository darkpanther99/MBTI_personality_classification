```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df = pd.read_csv('data/MBTI 500.csv')
```

```python
df
```

```python
pd.DataFrame.from_dict(Counter(df['posts'].str.split(' ').apply(len)), orient='index', columns=['Sequence lengths']).plot.bar()
```

```python
pd.DataFrame.from_dict(Counter(df['type']), orient='index', columns=['Occurance in dataset']).plot.bar()
```

```python
def convert_personality_type_to_binary(mbti_type):
    mapper = {
        'I':0,
        'E':1,
        'N':0,
        'S':1,
        'T':0,
        'F':1,
        'J':0,
        'P':1,
    }

    return [mapper[i] for i in mbti_type]
```

```python
df['binarized_types'] = df['type'].apply(convert_personality_type_to_binary)
```

```python
numpy_bin_types = np.vstack(df['binarized_types'].apply(np.array))
for i in range(4):
    pd.DataFrame.from_dict(Counter(numpy_bin_types[:, i]), orient='index', columns=['Occurance in dataset']).plot.bar()
```
